### Building Class that perform Simple LR 

```python
import pandas as pd
import numpy as np
```

```python
class MeraLR:
    def __init__(self):
        self.m =None
        self.b =None
    def fit(self,x_train,y_train):
        x_bar = x_train.mean()
        y_bar = y_train.mean()
        num = 0
        den = 0
        for i in range(x_train.shape[0]):
            num+=((y_train[i] - y_bar)*(x_train[i] - x_bar)) 
            den+=((x_train[i] - x_bar)*(x_train[i] - x_bar))
        self.m = num/den
        self.b = y_bar - (self.m * x_bar)
        print(self.m)
        print(self.b)

    def predict(self,x_test):
        y_result = [None] * x_test.shape[0]
        for _ in range(x_test.shape[0]):
           y_result[_] = ((self.m * x_test[_]) + self.b)
        return  y_result
```

```python
df = pd.read_csv("placement.csv")
df.head()
```

```python
x = df.iloc[:,0].values
y = df.iloc[:,1].values
```

```python
from sklearn.model_selection import train_test_split
x_train , x_test , y_train ,y_test = train_test_split(x,y,test_size=0.2,random_state=2)
```

```python
LR = MeraLR()
LR.fit(x_train,y_train)
```
