```python
# we Realised that Total Charges Should be Float , but it is object because there are blanks entries
# of just joined customer with tenure = 0 years
# blank --> 0
df['TotalCharges'] = df['TotalCharges'].replace(" ","0")
df['TotalCharges']= df['TotalCharges'].astype(float)
```

## Ploting

```python
ax = sns.countplot(x='Churn' , data=df)

for container in ax.containers:
    ax.bar_label(container)
plt.title('Count Of Customer By Churn')
plt.show()
```

```python
# Pie Chart
gb = df.groupby('Churn').count()
gb = df.groupby('Churn').agg({
    'Churn':'count'
})
gb
plt.figure(figsize=(3,4))
plt.pie(gb['Churn'],labels=gb.index , autopct='%1.1f%%')
plt.title('Percentage of Customer Churn')
plt.show()
```

```python
# Pie Chart
plt.figure(figsize=(4,4))
counts = df['Churn'].value_counts()

plt.pie(counts, labels=counts.index, autopct='%1.1f%%')
plt.title('Percentage of Customer Churn')
plt.show()
```

```python
# Exploring the Churn Analysis based on gender
plt.figure(figsize=(6,4))
sns.countplot(x='gender' , data = df , hue='Churn')
plt.title('Churn Based on Gender')
plt.show()
```

```python
plt.figure(figsize=(3,4))
ax = sns.countplot(x='SeniorCitizen', data=df ,hue='Churn')

for container in ax.containers:
    ax.bar_label(container)

plt.title('Senior Citizen Churn Analysis')
plt.show()
```

## Distribution of Customer and Churn based on tenure TimeLine

```python
plt.figure(figsize=(8,4))
sns.histplot(x='tenure' , data=df,bins=72,hue='Churn')
plt.title(" Distribution of Customer and Churn based on tenure TimeLine")
plt.show()
```

## Churn Analysis Based on Contract

```python
plt.figure(figsize=(5,3))
ax = sns.countplot(x='Contract', data=df ,hue='Churn')

for container in ax.containers:
    ax.bar_label(container)

plt.title('Contract Churn Analysis')
plt.show()
```
